```python
def relax(u, v, w):
    if w is not None and v.d > u.d + w:
        v.d = u.d + w
        v.p = u
```

```python
def bellman_ford_init(g, source):
    for v in g.v:
        v.d = sys.maxsize
        v.p = None
    source.d = 0
```

```python
def bellman_ford(g, source): # O(|V||E|)
    bellman_ford_init(g, source)
    for i in range(len(g.v)): # O(|V|)
        for u, v, w in g.edges(): # O(|E|)
            relax(u, v, w)
    for u, v, w in g.edges(): # O(|V|)
        if v.d > u.d + w:
            return False
    return True
```
